```python
import turtle
import os

wn = turtle.Screen()
wn.title("Pong by Deepank")
wn.bgcolor("black")
wn.setup(width = 800, height = 600)
wn.tracer(0)

# Score
score_a = 0
score_b = 0

# Paddle 1
paddle_a = turtle.Turtle()           # creating Turtle class
paddle_a.speed(0)                    # this is not the actual paddle speed but the speed of animation
paddle_a.shape("square")
paddle_a.color("white")
paddle_a.shapesize(stretch_wid=5, stretch_len=1)        # to increase the size of paddle
paddle_a.penup()
paddle_a.goto(-350,0)                # setting coordinate of paddle A to left

# Paddle 2
paddle_b = turtle.Turtle()           # creating Turtle class
paddle_b.speed(0)                    # this is not the actual paddle speed but the speed of animation
paddle_b.shape("square")
paddle_b.color("white")
paddle_b.shapesize(stretch_wid=5, stretch_len=1)        # to increase the size of paddle
paddle_b.penup()
paddle_b.goto(350,0)                # setting coordinate of paddle B to right

# Ball

ball = turtle.Turtle()           # creating Turtle class
ball.speed(0)                    # this is not the actual paddle speed but the speed of animation
ball.shape("square")
ball.color("white")
ball.penup()
ball.goto(0,0)                # setting coordinate of ball
ball.dx = 2                   # move by 2 pixels
ball.dy = 2

#pen
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()            # as we don't want the counter to display
pen.goto (0,260)
pen.write("Player A: 0 Player B: 0", align = "center", font=("Courier", 24, "normal"))

# Functions to move the paddles
def paddle_a_up():
    y = paddle_a.ycor()         # to get current y coordinate
    y += 20                     # adding 20 pixels to y
    paddle_a.sety(y)
    
def paddle_a_down():
    y = paddle_a.ycor()         # to get current y coordinate
    y -= 20                     # subtracting 20 pixels to y
    paddle_a.sety(y)    

def paddle_b_up():
    y = paddle_b.ycor()         # to get current y coordinate
    y += 20                     # adding 20 pixels to y
    paddle_b.sety(y)
    
def paddle_b_down():
    y = paddle_b.ycor()         # to get current y coordinate
    y -= 20                     # subtracting 20 pixels to y
    paddle_b.sety(y)
    
    
    
#keyboard Binding
wn.listen()
wn.onkeypress(paddle_a_up, "w")             # on key press w it would be moved up by 20 pixels
wn.onkeypress(paddle_a_down, "s")
wn.onkeypress(paddle_b_up, "Up")            # Up arrow Key
wn.onkeypress(paddle_b_down, "Down")

#Main game Loop
while True:
    wn.update()     #to update the screen everytime user run the program
    
    
# Move the ball
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy)
    
    
    
    #Border checking
    if ball.ycor() >290:             #to make it bounce of the top
        ball.sety(290)
        ball.dy *= -1w
        os.system("afplay bounce.wav&")          #for sound
        
    if ball.ycor() < -290:           #to make it bounce of the bottom
        ball.sety(-290)
        ball.dy *= -1
        os.system("afplay bounce.wav&")
        
    if ball.xcor() > 390:               #to check if the ball didnt hit the paddle
        ball.goto(0,0)
        ball.dx *= -1
        score_a += 1
        pen.clear()
        pen.write("Player A: {} Player B: {}".format(score_a,score_b), align = "center", font=("Courier", 24, "normal"))
        
    if ball.xcor()< -390:               #to check if the ball didnt hit the paddle
        ball.goto(0,0)
        ball.dx *= -1
        score_b += 1
        pen.clear()
        pen.write("Player A: {} Player B: {}".format(score_a,score_b), align = "center", font=("Courier", 24, "normal"))
        
        
    #paddle and ball collision
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor()+ 40 and ball.ycor() > paddle_b.ycor() -40):
        ball.setx(340)
        ball.dx *= -1
        os.system("afplay bounce.wav&")
        
    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor()+ 40 and ball.ycor() > paddle_a.ycor() -40):
        ball.setx(-340)
        ball.dx *= -1
        os.system("afplay bounce.wav&")
```
